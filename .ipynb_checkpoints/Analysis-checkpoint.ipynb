```python
mean=df['Pressure3pm'].mean()
mean

def impute_var(df,var):
    mean=df[var].mean()
    mode=df[var].mode()
    median=df[var].median()
    df[var+"_mean"]=df[var].fillna(mean)
    df[var+"_median"]=df[var].fillna(median)
    df[var+"_mode"]=df[var].fillna(mode[0])
    
    df[var+"_random"]=df[var]
    
    random_sample=df[var].dropna().sample(df[var].isnull().sum(),random_state=0)
    random_sample.index=df[df[var].isnull()].index
    df.loc[df[var].isnull(),var+'_random']=random_sample
    
impute_var(df,"Pressure3pm")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df['Pressure3pm'].plot(kind='kde', ax=ax)
df.Pressure3pm_mean.plot(kind='kde', ax=ax, color='red')
df.Pressure3pm_median.plot(kind='kde', ax=ax, color='yellow')
df.Pressure3pm_random.plot(kind='kde', ax=ax, color='green')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

```python
train_df=df.drop(["Date","Location","RainTomorrow"],axis=1)
```

```python
x=train_df.drop("RainTomorrow",axis=1)
# x=train_df
y=df["RainTomorrow"]
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
train_df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,f1_score,accuracy_score
conf=confusion_matrix(y_test,predictions)
```

```python
ascore=accuracy_score(y_test,predictions)
ascore
```
